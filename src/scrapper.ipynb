```python
import os
import requests
from bs4 import BeautifulSoup
```

```python
def get_story_links(main_url):
    response = requests.get(main_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    links = soup.find_all('a')
    story_links = [link.get('href') for link in links if 'author' in link.get('href')]
    return story_links
```

```python
def get_story_links_sub(main_url, name):
    response = requests.get(main_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    links = soup.find_all('a')
    story_links = [link.get('href') for link in links if name in link.get('href')]
    return story_links
```

```python
def scrape_story(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    paragraphs = soup.find_all('p')  # Assuming the story text is in <p> tags
    story_text = ' '.join(p.get_text().replace('\n', '') for p in paragraphs)
    return story_text
```

```python
def scrape_story2(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    paragraphs = soup.find_all('p')  # Assuming the story text is in <p> tags
    story_text = '\n\n\t'.join(p.get_text().strip() for p in paragraphs)
    return story_text
```

```python
def scrape_story3(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    paragraphs = soup.find_all('p')  # Assuming the story text is in <p> tags
    story_text = '\\n\\n\\t'.join(p.get_text().strip() for p in paragraphs)
    return story_text
```

```python
def save_story_as_text_file(story_name, story_text, file_directory):

    directory_path = f"{file_directory}/{story_name}"
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)

    file_path = f"{directory_path}/story.txt"
    with open(file_path, "w") as file:
        file.write(story_text)

    return file_path
```

```python
import os
import re  # Import the regular expressions module

# Clean fairy tales dataset
base_path = "./dataset/fairy tale"
for folder in os.listdir(base_path):
    if not folder.startswith('.DS_Store'):
        with open(os.path.join(base_path, folder, 'story.txt'), 'r') as file:
            file_contents = file.read()

        # Replace line breaks with spaces
        file_contents = file_contents.replace('\n', ' ')

        # Replace " with '
        story_text = file_contents.replace('"', "'")

        # Find the positions of fixed markers
        marker_position2 = story_text.find("Return to the The")
        marker_position3 = story_text.rfind("\n\n\t\n\n\t\n\n\t\n\n\t")

        # Find the position of the variable marker using regex for "Add .*? to your library"
        regex_pattern_add = r'Add .*? to your library'
        regex_match_add = re.search(regex_pattern_add, story_text)
        marker_position1 = regex_match_add.start() if regex_match_add else -1

        # Find the last position of "\n\n\t" and remove it along with following text
        marker_position4 = story_text.rfind("\n\n\t")

        # Find the position of the variable marker using regex for "<> was featured as <>"
        regex_pattern_featured = r'.*? was featured as .*'
        regex_match_featured = re.search(regex_pattern_featured, story_text)
        marker_position5 = regex_match_featured.start() if regex_match_featured else -1

        # Remove numbers at the end of the text
        story_text = re.sub(r'\s*\d+\.?\d*\s*$', '', story_text)

        # Create a list of valid positions (excluding -1)
        valid_positions = [pos for pos in [marker_position1, marker_position2, marker_position3, marker_position4, marker_position5] if pos != -1]
        
        # If any valid position is found, trim the content from the earliest marker
        if valid_positions:
            marker_position = min(valid_positions)
            story_text = story_text[:marker_position]

        # Create the output directory and write the cleaned text to a new file
        new_folder = os.path.join("./clean_dataset/fairy_tale", folder)
        os.makedirs(new_folder, exist_ok=True)
        new_file_path = os.path.join(new_folder, 'story.txt')
        
        with open(new_file_path, 'w') as new_file:
            new_file.write(story_text)
```
